```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Financial Columns

```python
dframe['sale_price_eur'].fillna(0, inplace=True)
dframe['monthly_rent_eur'].fillna(0, inplace=True)
```

Because missing values represent different listing types (sale vs rent), so 0 indicates not applicable rather than missing data

```python
# Step 1: city + property_type
dframe['bathrooms'] = dframe.groupby(['city', 'property_type'])['bathrooms']\
                    .transform(lambda x: x.fillna(x.median()))
# Step 2: city only
dframe['bathrooms'] = dframe.groupby('city')['bathrooms']\
                    .transform(lambda x: x.fillna(x.median()))
# Step 3: global median
dframe['bathrooms'].fillna(dframe['bathrooms'].median(), inplace=True)
```

```python
#square meters
dframe['square_meters'] = dframe.groupby(['city','property_type'])['square_meters']\
                        .transform(lambda x: x.fillna(x.median()))

dframe['square_meters'].fillna(dframe['square_meters'].median(), inplace=True)
```

```python
#year built
dframe['year_built'].fillna(dframe['year_built'].median(), inplace=True)
```

```python
#floor_number
dframe['floor_number'] = dframe.groupby(['city','property_type'])['floor_number']\
                       .transform(lambda x: x.fillna(x.median()))

dframe['floor_number'].fillna(0, inplace=True)
```

```python
#datys_on_market
dframe['days_on_market'] = dframe.groupby(['property_type','city'])['days_on_market']\
                         .transform(lambda x: x.fillna(x.median()))
dframe['days_on_market'] = dframe.groupby('property_type')['days_on_market']\
                         .transform(lambda x: x.fillna(x.median()))
dframe['days_on_market'].fillna(dframe['days_on_market'].median(), inplace=True)
```

```python
#price_per_sqm
dframe['price_per_sqm'] = dframe['sale_price_eur'] / dframe['square_meters']
```

```python
#furnishing_status
dframe['furnishing_status'].fillna("Unknown", inplace=True)
```

```python
#last_sold_price_eur
dframe['last_sold_price_eur'].fillna(dframe['last_sold_price_eur'].median(), inplace=True)
```

```python
#outliers
cols = [
    'sale_price_eur',
    'monthly_rent_eur',
    'square_meters',
    'price_per_sqm',
    'bathrooms',
    'days_on_market'
]

for col in cols:
    Q1 = dframe[col].quantile(0.25)
    Q3 = dframe[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    print(col, "outliers:", dframe[(dframe[col] < lower) | (dframe[col] > upper)].shape[0])
```

```python
#handling
Q1 = dframe['sale_price_eur'].quantile(0.25)
Q3 = dframe['sale_price_eur'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

dframe['sale_price_eur'] = dframe['sale_price_eur'].clip(lower, upper)
```

```python
Q1 = dframe['monthly_rent_eur'].quantile(0.25)
Q3 = dframe['monthly_rent_eur'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

dframe['monthly_rent_eur'] = dframe['monthly_rent_eur'].clip(lower, upper)
```

```python
Q1 = dframe['square_meters'].quantile(0.25)
Q3 = dframe['square_meters'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

dframe['square_meters'] = dframe['square_meters'].clip(lower, upper)
```

** SUMMARY OF DATA CLEANING PROCESS **
- Loaded the European real estate dataset and performed initial exploration:
  • Checked data types, missing values, and overall structure.
  • Identified key features such as pricing, property characteristics, and location data.

- Handled missing values using appropriate strategies:
  • Financial columns (sale_price_eur, monthly_rent_eur) filled with 0 to represent non-applicable listings (sale vs rent).
  • Numerical columns (bathrooms, square_meters, year_built, floor_number, etc.) imputed using median and group-based methods (city, property_type, bedrooms).
  • Categorical column (furnishing_status) filled with "Unknown".

- Applied feature-aware imputation:
  • Used hierarchical grouping (city, property_type, bedrooms) to preserve real-world patterns.
  • Avoided overly granular grouping such as latitude/longitude.

- Ensured data consistency:
  • Recalculated price_per_sqm from sale_price_eur and square_meters.
  • Verified logical consistency across related features.

- Checked for duplicates:
  • Verified both exact duplicates and duplicates based on property_id.
  • No significant duplicate issues found.

- Performed outlier detection using the IQR method:
  • Identified outliers in pricing and size-related columns.
  • Recognized that many outliers represent real-world variation (e.g., luxury properties).

- Applied outlier handling:
  • Used capping (winsorization) for:
      - sale_price_eur
      - monthly_rent_eur
      - square_meters
  • Retained realistic extreme values to avoid data loss.
  • No action needed for bathrooms and days_on_market as values were within reasonable ranges.

- Final validation:
  • Confirmed no missing values remain.
  • Ensured data is clean, consistent, and analysis-ready.

- Outcome:
  Prepared a high-quality dataset suitable for SQL-based analysis and Power BI visualization.
